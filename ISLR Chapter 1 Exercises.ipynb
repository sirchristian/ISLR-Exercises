```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 

df = pd.read_csv('datasets/College.csv', index_col=0)
df.head()
```

```python
df.describe()
```

```python
df.plot(kind='scatter', x='Apps', y='Accept', figsize=(12,8))  
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(df[df.columns[1:11]], figsize=(16, 18), diagonal='hist')
```

```python
df.boxplot(['Outstate'], by='Private', figsize=(8,6))
```

```python
df['Elite']=df.apply(lambda row: 'Yes' if row['Top10perc'] > 50 else 'No',axis=1)
df.where(df['Elite'] == 'Yes').describe()
```

```python
df.boxplot(['Outstate'], by='Elite', figsize=(8,6))
```
